```python
import psycopg2
```

```python
conn = psycopg2.connect("host=127.0.0.1 dbname=sparkifydb user=student password=student")
conn.set_session(autocommit=True)
cur = conn.cursor()
```

```python
cur.execute("""SELECT * FROM songplays
            WHERE song_id != 'None';""")
row = cur.fetchone()
while row:
    print(row)
    row = cur.fetchone()
```
